*Library Setup*

```python
# Library Imports
import os
import numpy as np
import matplotlib.pyplot as plt
import math
from skimage import io
from scipy import signal, ndimage

# Setup Jupyter notebook specifics
%matplotlib inline

# Numpy printing help
np.set_printoptions(precision=2)
```

**Part 1**

```python
# Create image array
ncols = 25
mrows = 25
J = np.ones((mrows,ncols))
f0 = 1

# Apply function to each pixel
for x in range(nrows):
    for y in range(mcols):
        J[y,x] = 20*math.sin(2*math.pi*f0*x) + 20*math.sin(2*math.pi*f0*y)

# Compute 2D FFT
J_FFT = np.fft.fft2(J)
# Magnitude + normalized
J_FFT_Mag = abs(J_FFT)*2/ncols*2/mrows
# Shift to only show one side of the FFT
J_FFT_Mag_shifted = np.fft.fftshift(J_FFT_Mag)

        
# Show the images
plt.subplots(2,2, figsize=(10,10))
plt.subplot(2,2,1); plt.imshow(J, cmap='gray'); plt.colorbar(); plt.title('Image')
plt.subplot(2,2,2); plt.hist(J.ravel(),bins=50); plt.xlabel('Intensity'); plt.ylabel('Count'); plt.title('Histogram')
plt.subplot(2,2,3); plt.imshow(J_FFT_Mag_shifted, cmap='gray'); plt.title('2D Fourier Transform')
plt.subplot(2,2,4).axis('off')
plt.show()
```
